### Import package and onnect to a SQL database.

```python
from os import getenv
from dotenv import load_dotenv
from mssql_python import connect

load_dotenv()
conn = connect(getenv("SQL_CONNECTION_STRING"))
```

### Execute a query

```python
SQL_QUERY = """
SELECT * FROM [dbo].[PerCapitaPersonalIncome];
"""

# Use cursor.execute to retrieve a result set from a query against the database.
cursor = conn.cursor()
cursor.execute(SQL_QUERY)

# Use cursor.fetchall with a foreach loop to get all the records from the database. Then print the records.

records = cursor.fetchall()
for r in records:
    # print the entire row; you can replace this with index-based access if desired
    print(r)
```
